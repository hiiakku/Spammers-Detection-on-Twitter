```python
import nltk
import pandas as pd
```

```python
twt_df = pd.read_csv('emozi_features.csv')
```

```python
from textblob import TextBlob
```

```python
nltk.download('punkt')
```

```python
nltk.download('averaged_perceptron_tagger')
```

```python
twt_sent = []
emozi_sent = []
for index,row in twt_df.iterrows():
    tb = TextBlob(str(row['tweet']))
    twt_sent.append(tb.sentiment.polarity)
    emozi_sent.append(2*row['aggregated_score']-1)
```

```python
twt_df['tweet_sentiment']=twt_sent
twt_df['emozi_sentiment'] = emozi_sent
```

```python
twt_df=twt_df.drop(twt_df.columns[[0]], axis=1)
twt_df.head(10)
```

```python
model_df = pd.read_csv('user_and_content_based_features.csv')
```

```python
model_df=model_df.drop(model_df.columns[[0]], axis=1)
model_df.head(10)
```

```python
from sklearn.cross_validation import train_test_split 
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
X = model_df.iloc[:, [1,2,3,4]].values
y = model_df.iloc[:, 5].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.36, random_state = 21)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
clf = SVC(kernel = 'rbf',C=90.0)
clf.fit(X_train, y_train)
```

```python
twt_data = pd.read_csv('live_content_based_features.csv')
twt_data=twt_data.drop(twt_data.columns[[0]], axis=1)
twt_data.head()
```

```python
x_test_twt = twt_data.iloc[: , [1,2,3,4]].values
```

```python
y_test_twt = clf.predict(x_test_twt)
```

```python
label = list(y_test_twt)
```

```python
twt_final_df = pd.DataFrame(
    {'user_id':twt_data['user_id'] ,
     'is_spam' : label
    })
```

```python
twt_final_df.head()
```

```python
df= pd.merge(twt_df, twt_final_df, on='user_id', how='inner')
```

```python
df.head(25)
```

```python
sentiment_df = []
for index,row in twt_df.iterrows():
    sentiment_df.append(abs(row['tweet_sentiment']-row['emozi_sentiment']))
    
df2 = pd.DataFrame(
    {'user_id':list(twt_df['user_id']) ,
     'sentiment_difference' : sentiment_df
    })
```

```python
final_df= pd.merge(twt_data, df2, on='user_id', how='inner')
final_df.head()
final_df = pd.merge(twt_final_df,final_df,on='user_id',how='inner')
final_df.head(50)
```

```python
X = final_df.iloc[:, [2,3,4,5,6]].values
y = final_df.iloc[:, 1].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.36, random_state = 21)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
clf = SVC(kernel = 'rbf',C=90.0)
clf.fit(X_train, y_train)
```

```python
print ("Testing Accuracy for svm is : ",accuracy_score(y_test,clf.predict(X_test)))
print ("Training Accuracy for svm is : ",accuracy_score(y_train,clf.predict(X_train)))
```
