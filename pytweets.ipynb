```python
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream

#Variables that contains the user credentials to access Twitter API 

access_token = "xxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
access_token_secret = "xxxxxxxxxxxxxxxxxxxxxxxxx"
consumer_key = "xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"
consumer_secret = "xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx"


class StdOutListener(StreamListener):

    def on_data(self, data):
        print (data, file=open("output1.txt", "a"))
        return True

    def on_error(self, status):
        print (status, file=open("output1.txt", "a"))


if __name__ == '__main__':

    #This handles Twitter authetification and the connection to Twitter Streaming API
    l = StdOutListener()
    auth = OAuthHandler(consumer_key, consumer_secret)
    auth.set_access_token(access_token, access_token_secret)
    stream = Stream(auth, l)

    #This line filter Twitter Streams to capture data by the keywords: 
    stream.filter(track=['#', 'http',  'https'])
```

```python
def CountFrequency(l1,s1): 
    freq = {} 
    sc = {}
    summ=0
    if(len(l1)==0):
        return 0
    for i in range(len(l1)): 
        item = l1[i]
        if (item in freq): 
            freq[item] += 1
            summ=summ+s1[i]
        else: 
            freq[item] = 1
            sc[item] = s1[i]
            summ=summ+s1[i];
    
    avg=summ/len(l1)
    
    
    for key, value in freq.items():
        if(value>=2 and value<=10 and sc[key]>0):
            avg=min(1,avg+0.1)
        elif(value>10 and sc[key]>0):
            avg=min(1,avg+0.2)
        elif(value>=2 and value<=10 and sc[key]<0):
            avg=max(-1,avg-0.1)
        elif(value>10 and sc[key]>0):
            avg=max(-1,avg-0.2)
        
    return avg
```

```python
emozi = []
emozi_des = []
emozi_sc = []
agg_sc = []
c = 0 
for index,row in df1.iterrows():
    emozi.append(0)
    emozi_des.append([])
    emozi_sc.append([])
    for x in str(row['text']):
        if(x in emozi_map):
            
            emozi_des[c].append( emozi_des_map[x])
            
            emozi_sc[c].append(emozi_map[x])
            
            emozi[c]=emozi[c] + 1
    agg_sc.append(CountFrequency(emozi_des[c],emozi_sc[c]))
    c = c + 1
```

```python
ids = list(df1['user'])
twt_id = list(df1['id'])
twt = list(df1['text'])

emozi_df = pd.DataFrame(
    {'user_id':ids ,
     'tweet id': twt_id,
     'tweet':twt ,
     'emozi_count': emozi ,
     'emozi_description':emozi_des,
     'emozi_scores': emozi_sc,
     'aggregated_score': agg_sc
    })
```
