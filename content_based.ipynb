```python
uid = 16195340
c = 0
df = twt_spam.head(100000)
df1 = pd.DataFrame(columns=twt_spam.columns)
for index,row in df.iterrows():
    if (row['user_id'] == uid):
        if c<20: 
           df1.loc[index] = row
           c = c + 1
    else:
       df1.loc[index] = row
       uid = row['user_id']
       c = 1
df1
```

```python
uid = 16195340
c = 0
lst_at = [0]
lst_links = [0]
lst_hash = [0]
#df = twt_spam.head(50000)
#df1 = pd.DataFrame(columns=twt_spam.columns)
for index,row in df1.iterrows():
    if (row['user_id'] == uid):
        lst_at[c] = lst_at[c] + (str)(row['text']).count('@')
        lst_links[c] = lst_links[c] + (str)(row['text']).count('http:') + (str)(row['text']).count('www') - (str)(row['text']).count('http://www')-(str)(row['text']).count('https://www')
        lst_hash[c] = lst_hash[c] + (str)(row['text']).count('#')
           
    else:
       lst_at.append(row['text'].count('@'))
       lst_links.append((str)(row['text']).count('http:') + (str)(row['text']).count('www') - (str)(row['text']).count('http://www')-(str)(row['text']).count('https://www'))
       lst_hash.append(row['text'].count('#'))
       uid = row['user_id']
       c = c + 1
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split 
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
X = final_df2.iloc[:, [1,2, 3]].values
y = final_df2.iloc[:, 4].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.36, random_state = 21)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
clf = SVC(kernel = 'rbf',C=90.0)
clf.fit(X_train, y_train)
```

```python
print ("Testing Accuracy for svm is : ",accuracy_score(y_test,clf.predict(X_test)))
print ("Training Accuracy for svm is : ",accuracy_score(y_train,clf.predict(X_train)))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
print ("Testing Accuracy for decision tree is : ",accuracy_score(y_test,clf.predict(X_test)))
print ("Training Accuracy for decision tree is : ",accuracy_score(y_train,clf.predict(X_train)))
```
