```python
import matplotlib.pyplot as plt
from sklearn.cross_validation import train_test_split 
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
```

```python
X = df.iloc[:, [1,2,3,4]].values
y = df.iloc[:, 5].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.36, random_state = 21)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
clf = SVC(kernel = 'rbf',C=90.0)
clf.fit(X_train, y_train)
```

```python
print ("Testing Accuracy for svm is : ",accuracy_score(y_test,clf.predict(X_test)))
print ("Training Accuracy for svm is : ",accuracy_score(y_train,clf.predict(X_train)))
```
